TASK#04 Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.

```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#load data
data = pd.read_csv('US_Accidents_March23_sampled_500k.csv')
```

```python
#data preprocessing
# Check for missing values
missing_values = data.isnull().sum()

# Remove duplicates
data.drop_duplicates(inplace=True)


# Check data types
data.info()
```

EDA

```python
# Descriptive statistics
summary_stats = data.describe()

# Visualize accident distribution by road conditions
plt.figure(figsize=(12, 6))
sns.countplot(x='State', data=data)
plt.xticks(rotation=45)
plt.xlabel('State')
plt.ylabel('Description')
plt.title('Accidents by Road Surface Type')
plt.show()

# Calculate correlations
# Note: Correlations are more suitable for numerical data, so we'll skip this for categorical fields.

# Visualize accident distribution by light conditions
plt.figure(figsize=(12, 6))
sns.countplot(x='No_Exit', data=data)
plt.xticks(rotation=45)
plt.xlabel('No_Exit')
plt.ylabel('Crossing')
plt.title('Accidents by Light Conditions')
plt.show()

# Visualize accident distribution by weather conditions
plt.figure(figsize=(12, 6))
sns.countplot(x='Junction', data=data)
plt.xticks(rotation=45)
plt.xlabel('Junction')
plt.ylabel('Weather')
plt.title('Accidents by Weather Conditions')
plt.show()

# Visualize accident distribution by accident severity
plt.figure(figsize=(12, 6))
sns.countplot(x='Amenity', data=data)
plt.xticks(rotation=45)
plt.xlabel('Amenity')
plt.ylabel('Bump')
plt.title('Accidents by Accident Severity')
plt.show()
```

```python
#contributing factors analysis
plt.figure(figsize=(12, 6))
sns.countplot(x='Give_Way', data=data)
plt.xticks(rotation=90)
plt.xlabel('Give_Way')
plt.ylabel('Precipitation')
plt.title('Accidents by Contributing Factors')
plt.show()
```
